```python
# celeba 도메인이 아닌 데이터를 도메인으로 사용하려면 ImageFolder 함수를 사용할 수 있도록 폴더를 설정해야 함.

path_ = sorted(glob.glob('/content/drive/MyDrive/데이터/new_album_image/trot/*.jpg'))

val_num = round(len(path_)*0.05)

import random
import shutil
import os

def split(img_list, val_count, train_path, val_path):
  # img_list : 이미지 경로 리스트
  # val_count : val에 할당할 이미지 개수
  # train_path : train 데이터 넣을 경로
  # val_path : val 데이터 넣을 경로
  

  # val 담을 이미지 리스트 저장
  val_files=[]
  for i in random.sample(img_list, val_count): # 랜덤으로 val_count만큼 이미지 img_list에서 고르기
    val_files.append(i)

  # 위에서 고르지 않은 이미지들을 train 담을 이미지 리스트로 저장
  train_files = [x for x in img_list if x not in val_files] # 리스트내포

  # 고른 이미지를 train_path, val_path폴더에 폭사
  for k in train_files:
    shutil.copy(k, train_path)
  
  for c in val_files:
    shutil.copy(c, val_path)

  print('train 폴더 이미지 개수 : {}\ntest 폴더 이미지 개수 : {}'.format(len(glob.glob(train_path+'/*')),len(glob.glob(val_path+'/*'))))

train_path = '/content/drive/MyDrive/개인별 실습/영현 - V.A.R./stargan2/StarGAN/data/RaFD/{}/{}'.format('train','trot')
os.makedirs(train_path, exist_ok=True)
test_path = '/content/drive/MyDrive/개인별 실습/영현 - V.A.R./stargan2/StarGAN/data/RaFD/{}/{}'.format('test','trot')
os.makedirs(test_path, exist_ok=True)

split(path_, val_num, train_path, test_path)
```

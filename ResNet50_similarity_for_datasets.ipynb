데이터셋에서 feature 추출

```python
# Function to extract features using ResNet50
def extract_features(img_path):
    img = keras.preprocessing.image.load_img(img_path, target_size=(224, 224))
    img_array = keras.preprocessing.image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = keras.applications.resnet50.preprocess_input(img_array)
    features = model.predict(img_array)
    features = np.reshape(features, (features.shape[0], -1))
    return features
```

```python
# Set up transform for preprocessing the images
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Create DataLoader to iterate over the images in batches
dataloader = torch.utils.data.DataLoader(dataset, batch_size=1, shuffle=False)

# Create a list of image paths from the ImageFolder dataset
image_paths = []
for batch_idx, (data, _) in enumerate(dataloader):
    for i in range(len(data)):
        image_paths.append(dataset.imgs[batch_idx * dataloader.batch_size + i][0])
```

```python
len(image_paths) #for check
```

```python
# Load images and extract features using ResNet50
features = []
for img_path in image_paths:
    feat = extract_features(img_path)
    features.append(feat)
features = np.vstack(features)
```

similarity scores

```python
most_similar = []

for i in input_features:
  similarity_scores = []
  for dataset_feature in features:
    score = np.dot(dataset_feature, i) / (np.linalg.norm(dataset_feature) * np.linalg.norm(i))
    similarity_scores.append(score)
  most_similar_index = np.argmax(similarity_scores)
  most_similar.append(most_similar_index)
```

input / 유사 이미지의 경로 pairing

```python
most_similar_path = []

for i in most_similar:
  path = image_paths[i]
  most_similar_path.append(path)
```

```python
dst_dir = 'path for save'

for i, filepath in enumerate(most_similar_path):
  image = Image.open(filepath)
  filename = f"name_for_save_image_file_{i+1}.jpg"
  dst_path = os.path.join(dst_dir, filename)
  image.save(dst_path)
```

```python
pairs = []

for input, similar in zip(input_paths, most_similar_path):
  pair = (input, similar)
  pairs.append(pair)
```

```python
for i, pair in enumerate(pairs):
  input_image = Image.open(pair[0])
  #input_image = input_image.transpose(Image.ROTATE_270)
  input_image = input_image.resize((1000,1000))
  similar_image = Image.open(pair[1])
  similar_image = similar_image.resize((1000,1000))

  new_image = Image.new('RGB', (input_image.width + similar_image.width, max(input_image.height, similar_image.height)))
  new_image.paste(input_image, (0, 0))
  new_image.paste(similar_image, (input_image.width, 0))

  new_image.save(f'path_and_name_for_save_result_image_file_{i+1}.jpg')
```
